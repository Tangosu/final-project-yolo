# Test

```python
# Set custom cfg to test mode 
%cd cfg
!sed -i 's/batch=64/batch=1/' yolov4-obj.cfg
!sed -i 's/subdivisions=16/subdivisions=1/' yolov4-obj.cfg
%cd ..
```

```python
# Running the detector with a test image. Thresh flag set at 0.3
!./darknet detector test data/obj.data cfg/yolov4-obj.cfg /mydrive/yolov4/backup/yolov4-obj_last.weights /mydrive/images/test4.PNG -thresh 0.3
download('predictions.jpg')
```

```python
# Test image
!./darknet detector test data/obj.data cfg/yolov4-obj.cfg /mydrive/yolov4/backup/yolov4-obj_last.weights /mydrive/images/test2.PNG -thresh 0.7
download('predictions.jpg')
```

```python
# Print JSON file for test image
!./darknet detector test data/obj.data cfg/yolov4-obj.cfg /mydrive/yolov4/backup/yolov4-obj_last.weights -ext_output -dont_show -out result.json /mydrive/images/test.PNG
download('result.json')
```
