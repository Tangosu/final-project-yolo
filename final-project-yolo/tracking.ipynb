```python
##working. can filter in array
current_arr = []
previous_arr = []
track_obj = {}
unique_obj = []
track_id = 0
name = ""

# Number of frames
start = 0
end = 52

# X coordinates
# LB = 0.18
# LR = 0.25
# C = 0.31
# RR = 0.37
# RB = 0.43
LB = 0.54
LR = 0.6
C = 0.66
RR = 0.7
RB = 0.77

while(start<end):
    start+=1
    print("frame",start)
    current_arr = []
    print(sorted_arr[start].get('filename'))

    # records name and coordinates of arrows on current frame
    for j in sorted_arr[start].get('objects'):
        if j.get('relative_coordinates').get('center_x') >= 0.5 and (j.get('relative_coordinates').get('center_y') <= 0.85 and j.get('relative_coordinates').get('center_y') >= 0.4):
            if j.get('relative_coordinates').get('center_x') >= LB and j.get('relative_coordinates').get('center_x') <= (LB + 0.04):
                name = "Left_Blue"
            if j.get('relative_coordinates').get('center_x') >= LR and j.get('relative_coordinates').get('center_x') < (LR + 0.06):
                name = "Left_Red"
            if j.get('relative_coordinates').get('center_x') >= C and j.get('relative_coordinates').get('center_x') < (C + 0.04):
                name = "Centre"
            if j.get('relative_coordinates').get('center_x') >= RR and j.get('relative_coordinates').get('center_x') < (RR + 0.07):
                name = "Right_Red"
            if j.get('relative_coordinates').get('center_x') >= RB and j.get('relative_coordinates').get('center_x') < (RB + 0.04):
                name = "Right_Blue"
            current_arr.append({'arrow':name, 
                                'x_coor':j.get('relative_coordinates').get('center_x'), 
                                'y_coor':j.get('relative_coordinates').get('center_y')})
    if len(track_obj) < 1:
#         # Compares the current y coordinates to the previous y coordinates.
#         # If the distance of the y coordinates are close (within 0.15), the current y coordinates will be
#         # added to the track_obj dictionary. This is used as a way to track a unique moving object
        for j in current_arr:
            for k in previous_arr:
                dist_x = k.get('x_coor') - j.get('x_coor')
                if abs(dist_x) <= 0.05: 
                    dist = k.get('y_coor') - j.get('y_coor')
                    print(1,abs(dist))
                    if abs(dist) < 0.15:
                        track_obj[track_id] = j
                        track_id += 1

    else:
        for j, k in track_obj.copy().items():
            # Same as above but if track_obj is not empty
            obj_exist = False
            for p in current_arr:
                dist_x = k.get('x_coor') - p.get('x_coor')
                print(4,track_obj.copy().items())
                print(3,p,k,abs(dist_x))
                if abs(dist_x) <= 0.05: 
                    dist = k.get('y_coor') - p.get('y_coor')
                    print('test', 'prev y coor',k.get('y_coor'), 'current y coor',p.get('y_coor'),abs(dist))
                    if abs(dist) < 0.15:
                        # If the dist is less 0.15, the object is already being tracked
                        track_obj[j] = p
                        obj_exist = True
            if not obj_exist:
                # Adds the object to unique_obj if the object does not exist
                print('Add this to unique object', k)
                track_obj.pop(j)
                unique_obj.append(k.get('arrow'))
                    
    print(current_arr)
    print(previous_arr)
    previous_arr = current_arr.copy()
```
